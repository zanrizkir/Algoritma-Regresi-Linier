```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from sklearn.linear_model import LinearRegression

# Membuat Tabel Hubungan Berat Ibu Hamil (Variabel Bebas) dengan Berat Bayi (Variabel Tak Bebas)
data = {
    'Berat_Ibu_Hamil': [49.4, 63.5, 68, 52.5, 54.4, 70.3, 50.8, 73.9, 65.8, 54.4],
    'Berat_Bayi': [3515, 3742, 3629, 2880, 3008, 4068, 3373, 4124, 3572, 3359]
}

# Mengubah data menjadi DataFrame Pandas (Tabel)
df = pd.DataFrame(data)
df
```

```python
# X adalah fitur/prediktor (Berat Badan Ibu Hamil)
X = df[['Berat_Ibu_Hamil']]
# y adalah target yang ingin diprediksi (Berat Badan Bayi yang akan Dilahirkan)
y = df['Berat_Bayi']

# Inisialisasi model
model = LinearRegression()
# Melatih model dengan data tabel yang sudah dibuat
model.fit(X, y)

# Visualising the Training set results
plt.scatter (X, y, color = 'red')
plt.plot(X, model.predict(X), color = 'blue')
plt.title('Berat Ibu Hamil vs Berat Bayi yang Dilahirkan')
plt.xlabel('Berat Ibu Hamil (Kg)')
plt.ylabel('Berat Bayi (gram)')
plt.show()
```

```python
# Prediksi untuk berat badan bayi yang akan dilahirkan
data_baru = pd.DataFrame({'Berat_Ibu_Hamil': [80, 85, 100]})
prediksi = model.predict(data_baru)

# Menampilkan hasil prediksi
for x_val, y_pred in zip(data_baru['Berat_Ibu_Hamil'], prediksi):
    print(f"Prediksi Berat Badan Bayi untuk Ibu hamil {x_val} kg: Rp {y_pred:.3f} gram")
```
